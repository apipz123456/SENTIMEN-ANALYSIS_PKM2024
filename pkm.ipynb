```python
def tokenize_text(text):
    return word_tokenize(text)

def clean_text(text):
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    return text


komentar_pondok['tokens'] = komentar_pondok['komentar'].apply(tokenize_text)
komentar_pondok['tokens'] = komentar_pondok['tokens'].apply(lambda tokens: [clean_text(token) for token in tokens])
print(komentar_pondok[['tokens']])
```
